# finding the duplicates and removing the duplicates

```python
import numpy as np
import pandas as pd
```

```python
# read the data from the csv file 
df=pd.read_csv("C:\\Users\\dell\\pandas\\studentmarks1.csv")
print(df)
```

```python
# finding the duplicates in this dataframe 
df.duplicated()
```

```python
# this is giving the boolean result
```

```python
# finding the number of duplicate records 
df.duplicated().sum()
```

```python
# find the duplicated record from the dataframe
df[df.duplicated()]
```

```python
# removing the duplicated record
df.drop_duplicates()
```

```python
# here we rae drop the duplicate record
# but it is not perminantly removed 
# for perminante removing purpose we are using the inplace=True|
```

```python
df
```

```python
# 12 record is available 
# now we are removing perminantly
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
# 12 record is perminantly removed from the dataframe 
```
