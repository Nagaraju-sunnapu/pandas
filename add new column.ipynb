# Add new column,filtering the data and updatation

```python
# read the data from the csv file 
import numpy as np
import pandas as pd
df=pd.read_csv("C:\\Users\\dell\\pandas\\studentmarks1.csv")
print(df)
```

```python
# adding the new column on the object of dataframe with the name of "total"
df["total"]=df["telugu"]+df["english"]+df["hindi"]+df["maths"]+df["science"]+df["social"]
```

```python
print(df)
```

```python
# add the new column on the object of dataframe with name of te column is "percent"
df["percent"]=(df["total"]/600)*100
```

```python
df["percent"]=round((df["total"]/600)*100,2)
```

```python
print(df)
```

```python
# datafiltering and conditional change
df[df["percent"]>70]
```

```python
# get the maths marks who are getting 75-95 
df[(df["maths"]>=75) & (df["maths"]<=95)]
```

```python
# access the name and maths marks who are getting maths marks in between 75-95 
df.loc[(df["maths"]>=75) & (df["maths"]<=95)][["name","maths"]]
```

# data filtering and updating the values

```python
# data filtering and updating the values 
# assingning the grades
# DISTINCTION--> provided the percentage >=72
# FIRST--> provided persent ranges between less than 72 and greater than or equal to 65
# SECOND--> provided persent ranges between less than 65 and greater than or eqaul to 60
# THIRD--> provided percent less than 60
```

```python
df["grade"]=None
```

```python
print(df)
```

```python
df.loc[df["percent"]>=72,["grade"]]="DISTINCTION"
```

```python
print(df)
```

```python
# FIRST--> provided persent ranges between less than 72 and greater than or equal to 65
df.loc[(df["percent"]<72) & (df["percent"]>=65),["grade"]]="FIRST"
```

```python
print(df)
```

```python
# SECOND--> provided persent ranges between less than 65 and greater than or eqaul to 60
df.loc[(df["percent"]<65) & (df["percent"]>=60),["grade"]]="SECOND"
```

```python
print(df)
```

```python
# THIRD--> provided percent less than 60
df.loc[df["percent"]<60,["grade"]]="THIRD"
```
