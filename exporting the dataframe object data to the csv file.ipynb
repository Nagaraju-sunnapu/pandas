# exporting the dataframe object data to the csv and excel files 

```python
import numpy as np
import pandas as pd
```

```python
# read the data from the csv file 
df=pd.read_csv("C:\\Users\\dell\\pandas\\studentmarks1.csv")
print(df)
```

```python
# add the new column total
df["total"]=df["telugu"]+df["english"]+df["hindi"]+df["maths"]+df["science"]+df["social"]
print(df)
```

```python
# add another column percent
df["percent"]=round((df["total"]/600)*100,2)
```

```python
df
```

```python
# assigning grades
# DISTINCTION-->provided percent >=72
# "FIRST"-->provided percent ranges between less than 72 and grater than or equal to 65
# "SECOND"-->provided percent ranges between less than 65 and grater than or equal to 60
# "THIRD"--> provided percent less than 60
```

```python
df.loc[df["percent"]>=72,["grade"]]="DISTINCTION"
df
```

```python
df.loc[(df["percent"]<72) & (df["percent"]>=65),["grade"]]="FIRST"
df
```

```python
# "SECOND"-->provided percent ranges between less than 65 and grater than or equal to 60
df.loc[(df["percent"]<65) & (df["percent"]>=60),["grade"]]="SECOND"
df
```

```python
# "THIRD"--> provided percent less than 60
df.loc[df["percent"]<60,["grade"]]="THIRD"
df
```

```python
# exporting the dataframe object data to the csv file 
df.to_csv("C:\\Users\\dell\\pandas\\studentmarksreport.csv")
print("dataframe data exported successfully to the csv file")
```

```python
# exporting the dataframe object data into excel file 
df.to_excel("C:\\Users\\dell\\pandas\\studentsfinalmarks.xlsx")
print("dataframe object data successfully enter to the excel file")
```
